```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("/content/Maternal Health Risk Data Set.csv")
```

```python
 # Remove duplicates
 df.drop_duplicates()
```

```python
# Scatter Plot: Age vs SystolicBP
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["Age"], y=df["SystolicBP"], hue=df["RiskLevel"], palette="coolwarm")
plt.title("Scatter Plot: Age vs SystolicBP")
plt.xlabel("Age")
plt.ylabel("SystolicBP")
plt.show()
```

```python
# Bar Plot: Risk Level count
plt.figure(figsize=(6,4))
sns.countplot(x=df["RiskLevel"], palette="viridis")
plt.title("Bar Plot: Risk Level Distribution ")
plt.xlabel("Risk Level")
plt.ylabel("Count")
plt.show()
```

```python
# Box Plot: Distribution of Blood Sugar by Risk Level
plt.figure(figsize=(6,4))
sns.boxplot(x=df["RiskLevel"], y=df["BS"], palette="pastel")
plt.title("Box Plot: Blood Sugar Levels by Risk Level")
plt.xlabel("Risk Level")
plt.ylabel("Blood Sugar Level")
plt.show()
```

```python
# Pie Chart: Risk Level distribution
risk_counts = df["RiskLevel"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(risk_counts, labels=risk_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen'])
plt.title("Pie Chart: Risk Level Distribution")
plt.show()
```

```python
# Line Chart: Age vs HeartRate
plt.figure(figsize=(6,4))
sns.lineplot(x=df["Age"], y=df["HeartRate"], marker="o", color="purple")
plt.title("Line Chart: Age vs Heart Rate ")
plt.xlabel("Age")
plt.ylabel("Heart Rate")
plt.show()
```
